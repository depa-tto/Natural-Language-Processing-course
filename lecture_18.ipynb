```python
from scipy import spatial
```

```python
cosine_similarity = lambda vec1, vec2: 1 - spatial.distance.cosine(vec1,vec2)
```

```python
king = nlp.vocab['king'].vector
```

```python
man = nlp.vocab['man'].vector
```

```python
woman = nlp.vocab['woman'].vector
```

```python
# ask for the king - man + woman ---> new vecror that will be similar to queen, princess or highness
```

```python
new_vector = king-man+woman
```

```python
computed_similarity = []

for word in nlp.vocab:
    if word.has_vector:
        if word.is_lower:
            if word.is_alpha:
                similarity = cosine_similarity(new_vector,word.vector)
                computed_similarity.append((word,similarity))
```

```python
computed_similarity = sorted(computed_similarity,key=lambda item:-item[1]) # sroting acording to similarity in descending order
```

```python
print([t[0].text for t in computed_similarity[:10]])
```
