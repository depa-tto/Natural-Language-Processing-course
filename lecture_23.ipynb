```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import keras
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
y = iris.target
```

```python
from keras.utils import to_categorical # type:ignore
```

```python
y = to_categorical(y)
```

```python
y.shape # 3 vaues for label
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_object = MinMaxScaler()
```

```python
scaler_object.fit(X_train)
```

```python
scaled_X_train = scaler_object.transform(X_train)
```

```python
scaled_X_test = scaler_object.transform(X_test)
```

```python
from keras.models import Sequential # type:ignore
from keras.layers import Dense # type:ignore
```

```python
model = Sequential()
model.add(Dense(8,input_dim=4,activation='relu')) # dense layer to the model
model.add(Dense(8,input_dim=4,activation='relu')) # another set of neurons
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(scaled_X_train,y_train,epochs=150,verbose=2)
```

```python
model.predict(scaled_X_test)
```

```python
y_prob = model.predict(scaled_X_test)
predictions = np.argmax(y_prob, axis=1)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(classification_report(y_test.argmax(axis=1),predictions))
```

```python
confusion_matrix(y_test.argmax(axis=1),predictions)
```

```python
accuracy_score(y_test.argmax(axis=1),predictions)
```
