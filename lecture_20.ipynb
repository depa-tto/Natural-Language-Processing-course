```python
import numpy as np
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('./datasets_files/moviereviews.tsv',sep='\t')
```

```python
blanks = []

for i,lb,rv in df.itertuples(): # for index label and review
    if type(rv) == str:
        if rv.isspace():
            blanks.append(i)
```

```python
df.dropna(inplace=True)
```

```python
blanks
```

```python
df.drop(blanks,inplace=True)
```

```python
df['label'].value_counts()
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
df['scores'] = df['review'].apply(lambda review:sid.polarity_scores(review))
```

```python
df['compound'] = df['scores'].apply(lambda d:d['compound'])
```

```python
df['comp_score'] = df['compound'].apply(lambda score: 'pos' if score >= 0 else 'neg')
```

```python
df.head()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(df['label'], df['comp_score']) 
```

```python
print(classification_report(df['label'], df['comp_score'])) # again problem with sarcasm
```

```python
confusion_matrix(df['label'], df['comp_score'])
```
